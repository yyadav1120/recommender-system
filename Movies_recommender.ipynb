```python
import pandas as pd
import numpy as np
import json
```

```python
def read_json(filename: str) -> dict:
  
    try:
        with open(filename, "r") as f:
            data = json.loads(f.read())
    except:
        raise Exception(f"Reading {filename} file encountered an error")
  
    return data
  
```

```python
def create_dataframe(data: list) -> pd.DataFrame:
  
    # Declare an empty dataframe to append records
    dataframe = pd.DataFrame()
  
    # Looping through each record
    for d in data:
          
        # Normalize the column levels
        record = pd.json_normalize(d)
          
        # Append it to the dataframe 
        dataframe = dataframe.append(record, ignore_index=True)
  
    return dataframe
```

```python
data = read_json(filename="dataset.json")
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
vectorizer = TfidfVectorizer()
```

```python
item_ids = df['id']
item_titles = df['title']
item_features = df.drop(['id', 'title'], axis=1).astype(str)
item_features['combined_features'] = item_features.apply(lambda x: ' '.join(x), axis=1)
vectorizer = TfidfVectorizer()
item_features_matrix = vectorizer.fit_transform(item_features['combined_features'])
cosine_sim_matrix = cosine_similarity(item_features_matrix, item_features_matrix)
```

```python
def get_recommendations(item_title, top_n=5):
    item_index = item_titles[item_titles == item_title].index[0]
    sim_scores = list(enumerate(cosine_sim_matrix[item_index]))
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)
    top_items = [(item_ids[i], item_titles[i], score) for i, score in sim_scores if i != item_index][:top_n]
    return top_items
```

```python
recommendations = get_recommendations('Heat', top_n=5)
print(recommendations)
```
